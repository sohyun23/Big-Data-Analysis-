### 문제1

제공된 데이터셋(df)의 age(나이) 컬럼을 활용하여 제3사분위수(75% 백분위수)와 제1사분위수(25% 백분위수)의 차이(IQR: 사분위간 범위)를 구하시오. 계산된 결과의 소수점 이하를 버리고 정수 형태로 제출하시오.

```python
import pandas as pd
# df = pd.read_csv("data4-1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-1.csv")

# 1) 3사분위수와 1사분위수의 차이를 절대값으로 계산
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
result = abs(q1 - q3)

# 2) 소수점 이하를 버리고, 정수로 출력
print(int(result))
```

### 문제2

제공된 데이터셋(df)에서 아래 3가지 조건을 모두 만족하는 데이터(행)의 총 개수를 구하시오.

(loves + wows) / reactions의 비율이 40% 초과

(loves + wows) / reactions의 비율이 50% 미만

type 컬럼의 값이 'video'인 데이터

```python
import pandas as pd
# df = pd.read_csv("data4-2.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-2.csv")
df
```

```python
df['ratio'] = (df['loves'] + df['wows'])/df['reactions']
cond1 = df['ratio'] > 0.4
cond2 = df['ratio'] < 0.5
cond3 = df['type'] == 'video'

print(len(df[cond1 & cond2 & cond3]))
```

```python
import pandas as pd
# df = pd.read_csv("data4-2.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-2.csv")

# 1) 비율이 40%보다 크고 50%보다 작은 조건
cond1 = (df['loves'] + df['wows']) / df['reactions'] > 0.4
cond2 = (df['loves'] + df['wows']) / df['reactions'] < 0.5

# 2) type이 video인 조건
cond3 = df['type'] == 'video'

# 3) 조건에 맞는 데이터 수 구하기
print(len(df[cond1 & cond2 & cond3]))
```
