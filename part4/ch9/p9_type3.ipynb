### 문제1-1

데이터를 분할하고 회귀 모델을 구축한 뒤 유의한 변수의 개수를 구하시오

데이터를 id <= 140은 train, id > 140은 test 데이터로 분할하시오. train 데이터를 사용하여 design을 종속변수로, c1~c5를 독립변수로 하는 다중선형회귀분석을 수행하고, 상수항을 제외한 독립변수 중 유의수준 0.05 미만인 유의한 변수의 개수를 구하시오.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/design.csv")
df
```

```python
train = df[df['id'] <= 140]
print(train)
test = df[df['id'] > 140]
```

```python
import statsmodels.formula.api as smf
formula = "design ~ c1 + c2 + c3 + c4 + c5"
model = smf.ols(formula, data = train).fit()
print(model.summary())
```

```python
print((model.pvalues[1:] < 0.05).sum())
```

```python
import pandas as pd
df = pd.read_csv("design.csv")
# df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/design.csv")
df.head()
```

```python
from statsmodels.formula.api import ols

# 1) train, test 데이터 분리
cond1 = df['id'] <= 140
cond2 = df['id'] > 140
train = df[cond1].copy()
test = df[cond2].copy()

# 2) 전체 변수 사용 회귀분석
model = ols("design ~ c1 + c2 + c3 + c4 + c5", data=train).fit()
print(model.summary())

# 3) 상수항(Intercept)은 제외하고, p-value가 0.05보다 작은 독립변수의 개수를 계산
print(model.pvalues[1:] < 0.05)
sum(model.pvalues[1:] < 0.05)
```

### 문제1-2

[문제 1]에서 선택된 유의한 독립변수들(c1, c2, c4)만 사용하여 train 데이터로 회귀 모델을 재학습시키시오. 이 모델을 통해 얻은 train 데이터의 design 예측값과 실제 design 값 간의 피어슨 상관계수를 구하시오.

출력 형식: 소수점 셋째 자리까지 반올림 (예: 0.785)

```python
formula2 = "design ~ c1+ c2+ c4"
model = smf.ols(formula2, data = train).fit()
pred = model.predict(train)
print(pred)
```

```python
df_design = pd.DataFrame({'train' : train['design'], 'pred': pred})
df_design
```

```python
#df_design.corr().loc['train', 'pred']

round(train['design'].corr(pred),3) # 피어슨이 default
print(round(train['design'].corr(pred),3))
```

```python
# 방법1
# 1) 유의한 변수만 사용한 회귀 분석
model = ols("design ~ c1 + c2 + c4", data=train).fit()

# 2) train 데이터에서 design의 예측
train['pred_design'] = model.predict(train)

# 3) 피어슨 상관계수 계산
result = train['design'].corr(train['pred_design'])
round(result,3)
```

```python
# 방법2(추가tip)
from scipy.stats import pearsonr
# 1) 유의한 변수만 사용한 회귀 분석
model = ols("design ~ c1 + c2 + c4", data=train).fit()

# 학습데이터에서 예측값 계산
train['pred_design'] = model.predict(train)

# 예측값과 실제 design 값 사이의 피어슨 상관계수 계산
c, p = pearsonr(train['design'], train['pred_design'])
round(c,3)
```

```python
test
```

```python
prediction = model.predict(test)

import numpy as np
from sklearn.metrics import mean_squared_error
def rmse(y, y_pred):
  return np.sqrt(mean_squared_error(y,y_pred))

print(rmse(test['design'], prediction))
```
