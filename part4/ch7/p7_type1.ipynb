### 문제3

제공된 데이터셋(df)의 CO2 컬럼을 대상으로 IQR(사분위간 범위) 방식에 기반한 이상치 검출을 수행하고자 한다. 아래의 이상치 판단 기준(상한값 초과 또는 하한값 미만)을 적용했을 때, 이상치에 해당하는 데이터(행)의 총 개수를 정수 형태로 제출하시오.

$\text{IQR} = \text{제3사분위수(Q3)} - \text{제1사분위수(Q1)}$$\text{상한선(Upper Boundary)} = \text{Q3} + 1.5 \times \text{IQR}$$\text{하한선(Lower Boundary)} = \text{Q1} - 1.5 \times \text{IQR}$

```python
Q1 = df['CO2'].quantile(0.25)
Q3 = df['CO2'].quantile(0.75)
IQR = Q3- Q1
print(Q1, Q3, IQR)
upper = Q3 + 1.5*IQR
lower = Q1 - 1.5*IQR
print(len(df[(df['CO2']> upper) | (df['CO2'] < lower)]))
```

```python
import pandas as pd
# df = pd.read_csv("air_quality.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/air_quality.csv")

# 1) IQR 계산
Q1 = df['CO2'].quantile(0.25)
Q3 = df['CO2'].quantile(0.75)
IQR = Q3 - Q1

# 2) 상한 및 하한 계산
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

# 3) 이상치 식별
outliers = df[(df['CO2'] < lower) | (df['CO2'] > upper)]

# 4) 이상치 수
print(len(outliers))
```
