### 문제1-3

[문제 1-1]에서 학습시킨 단일 변수(weight) 로지스틱 회귀 모델을 사용하되, test 데이터의 gender 값을 예측하시오. 예측된 확률 값이 0.5를 초과(이상)하는 경우 1, 그렇지 않은 경우 0으로 분류했을 때의 오류율을 구하시오

출력형식 셋째자리

```python
test = df[210:]

import statsmodels.formula.api as smf
formula = "gender ~ weight "
model = smf.logit(formula, data = train).fit()
pred = model.predict(test)
print((pred > 0.5).astype(int))
```

```python
from sklearn.metrics import accuracy_score
prediction = (pred > 0.5).astype(int)
error_rate = 1 - accuracy_score(test['gender'], prediction)
print(round(error_rate,3))
```

```python
from sklearn.metrics import accuracy_score

# 1) test데이터를 사용해 예측 (0.5 미만: 0, 0.5 이상 1)
model = logit("gender ~ weight", data=train).fit()
target = test.pop('gender')
pred = model.predict(test)
pred = (pred > 0.5).astype(int)

# 2) 실제 값과 예측값을 사용하여 정확도 계산
accuracy = accuracy_score(target, pred)

# 3) 오류율 계산
error_rate = 1 - accuracy
print(round(error_rate, 3))
# 0.478
```

### 문제2-2, 2-3

문제 2-2]
데이터셋에서 CPU 변수의 값이 100 미만인 데이터만 필터링하여 분석을 진행하시오. 필터링된 데이터를 바탕으로 ERP를 종속변수로 하고, Feature1, Feature2, Feature3, CPU를 독립변수로 설정하여 다중선형회귀분석을 수행하시오. 분석 결과 중 회귀 모델의 요약 정보(Summary)를 확인하여 문제를 해결하시오.
(주로 이 단계에서 특정 변수의 회귀계수, p-value, 또는 결정계수 등을 최종 값으로 요구합니다.)

```python
df1 = df[df['CPU'] < 100]
df1

import statsmodels.formula.api as smf
formula = "ERP ~ Feature1 + Feature2 + Feature3 + CPU"
model = smf.ols(formula, data = df1).fit()
print(model.summary())
```

```python
from statsmodels.formula.api import ols

# 1) CPU가 100 미만인 데이터 필터링
filtered_df = df[df['CPU'] < 100]

# 2) 선형회귀 모델 생성: ERP를 종속 변수로, 나머지 변수들을 독립 변수로 설정
model = ols('ERP ~ Feature1 + Feature2 + Feature3 + CPU', data=filtered_df).fit()

# 3) 모델 요약 정보 출력
print(model.summary())
```
