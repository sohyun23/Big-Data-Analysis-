### 문제1-1

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/attrition.csv")

# 1) 로지스틱 회귀분석 수행
from statsmodels.formula.api import logit
model = logit("attrition ~ age + income + overtime", data=df).fit()

# 2) 회귀 분석 결과 출력
print(model.summary())

# 3) income 회귀계수 출력
print(round(model.params["income"], 3))

# 정답: -0.002
```

정답은 소수점 넷째 자리에서 반올림하여 소수점 셋째 자리까지 출력하시오

```python
import numpy as np
print(round(np.exp(model.params['age']),3))
```

```python
# 1) 오즈비 계산
import numpy as np
result = np.exp(model.params["age"])

# 2) 반올림 후 결과 출력
print(round(result, 3))

# 정답: 0.996
```

### 문제2-1

2-1. 회귀계수 계산 및 합 구하기

독립변수로 외벽 면적(wall), 지붕 면적(roof), 유리창 면적(glazing), 건물의 높이(height)를 사용하여 다중 선형 회귀모형을 적합한 후, 절편(Intercept)을 제외한 모든 독립변수들의 회귀계수(Coefficient)의 합을 구하시오.

정답은 소수점 넷째 자리에서 반올림하여 소수점 셋째 자리까지 출력하시오.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
print(dir(smf))
```

```python
# 1) 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/heating.csv")
df
```

```python
formula2 = "heating_load ~ wall + roof + glazing + height"
model2 = smf.ols(formula2, data = df).fit()
#print(model2.summary())
print(round(model2.params[1:].sum(),3))
```

```python
import pandas as pd

# 1) 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch10/heating.csv")

# 2) 다중선형회귀분석
from statsmodels.formula.api import ols
model = ols("heating_load ~ wall + roof + glazing + height", data=df).fit()

# 3) 회귀 분석 결과 출력
print(model.summary())

# 4) 절편 제외 회귀계수들의 합 구하기
result = model.params[1:].sum()
print(round(result, 3))

# 정답: 0.254
```

### 문제2-2

2-2. 변수 제거 후 결정계수($R^2$)
 확인하기

 기존 모형에서 독립변수 중 외벽 면적(wall) 변수를 제거하고, 종속변수 heating_load에 대해 나머지 3개 변수(roof, glazing, height)만을 사용하여 다중 선형 회귀모형을 다시 적합하시오. 이때 변경된 모형의 **결정계수(R-squared)**를 구하시오.

 정답은 소수점 넷째 자리에서 반올림하여 소수점 셋째 자리까지 출력하시오.

```python
new_formula2 = "heating_load ~ roof + glazing + height"
new_model2 = smf.ols(new_formula2, data = df).fit()
print(new_model2.summary())
print(round(new_model2.rsquared,3))
```

```python
# 1) 다시 회귀 분석
model = ols("heating_load ~ roof + glazing + height", data=df).fit()

# 2) 결정계수(R-squared) 확인
print(model.summary())

# 정답: 0.754
```
