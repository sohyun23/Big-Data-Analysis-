### 문제1

어느 병원에서 항암약을 투여받은 환자들의 부작용 결과를 조사한 데이터(df)가 있다. 부작용의 범주는 총 4개(1, 2, 3, 4)로 분류되며, 4는 '이상 없음'을 의미한다. 기존 연구에 따르면 각 부작용 범주의 기대 비율은 1: 10%, 2: 5%, 3: 15%, 4: 70%라고 한다.

[1-1] 항암약을 투여받은 전체 환자 중 부작용 결과가 '이상 없음(4)'인 환자의 비율을 구하시오.출력 형식: 소수점 셋째 자리까지 반올림 (예: 0.750)

[1-2] 기존 연구의 기대 비율을 바탕으로 귀무가설($H_0$)에 대한 카이제곱 적합도 검정을 수행하고, 카이제곱 통계량(Statistic)을 구하시오.출력 형식: 소수점 둘째 또는 셋째 자리까지 반올림 (예: 2.345)

[1-3] 위 카이제곱 검정 결과의 p-value(유의확률)를 구하고, 유의수준 5%($\alpha=0.05$) 하에서 귀무가설의 기각 여부를 결정하시오.출력 형식: p-value 수치 및 기각 여부 선택형

```python
# 1-1
print((df['항암약'] == 4).sum()/ len(df))
```

```python
# 1-2.
from scipy import stats
#print(len(df))
total = len(df)
# 기대빈도수 계산
expt = [0.1 *total , 0.05 * total, 0.15 * total, 0.7* total ]
#print(help(stats.chisquare))

# 관찰빈도수 계산
obs = df.value_counts().sort_index(ascending = True).to_list()

# 카이제곱 적합도 검정
statistic, p_value = stats.chisquare(obs, expt)
print(statistic)
print(p_value) # 귀무가설 채택
```

```python
import pandas as pd
df = pd.DataFrame({
    "항암약":[4,4,3,4,1,4,1,4,1,4,4,2,1,4,2,3,2,4,4,4]
    })
print(df.head(3))
```

```python
# 방법1)
# 1) 이상 없음(4)의 빈도 계산
cnt = sum(df['항암약']==4)

# 2) 항암약을 투여 받은 환자 중 '이상 없음' 비율 계산
ratio = cnt / len(df)
print(ratio)

# 방법2)
# 1) 카테고리별 비율 계산
print(df['항암약'].value_counts(normalize=True))
```

```python
from scipy.stats import chisquare

# 1) 각 카테고리의 비율을 리스트로 만들기
prob = [0.1, 0.05, 0.15, 0.7]

# 2-1) 기대 빈도수 계산
print("데이터 수: ", len(df))
expected_counts = [0.1*20, 0.05*20, 0.15*20, 0.7*20]
print(expected_counts)

# 2-2)기대 빈도수 계산 (다른 방법)
expected_counts = [x*len(df) for x in prob]
print(expected_counts)
```

```python
# 3) 관찰 빈도수 계산
observed_counts = df['항암약'].value_counts().sort_index().to_list()
print(observed_counts)
```

```python
# 4) 카이제곱 검정 수행
print(chisquare(f_obs=observed_counts, f_exp=expected_counts))
```

```python
# 참고 - f_obs, f_exp 생략 가능
print(chisquare(observed_counts, expected_counts))
```

### 문제2

기상 관측 데이터(df)를 활용하여 기온(temperature)에 영향을 미치는 요인들을 분석하고자 한다. 기온을 종속변수(Y)로 하고, 일사량(solar), 풍속(wind), 오존농도(o3)를 독립변수(X)로 하는 다중선형회귀분석을 수행한 뒤 아래 질문에 답하시오.

[2-1] 독립변수 o3(오존농도)의 회귀계수(Coefficient)를 구하시오.

출력 형식: 소수점 셋째 자리까지 반올림 (예: 0.123)

[2-2] 독립변수 wind(풍속)의 유의확률(p-value)을 구하시오.

출력 형식: 소수점 셋째 자리까지 반올림 (예: 0.004)

[2-3] 학습된 회귀 모델을 바탕으로 solar = 100, wind = 5, o3 = 30일 때의 기온(temperature) 예측값을 구하시오.

출력 형식: 소수점 셋째 자리까지 반올림 (예: 24.567)

```python
import pandas as pd
# df = pd.read_csv("data6-3-2.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch6/data6-3-2.csv")

print(df.head(3))
```

```python
import statsmodels.formula.api as smf

formula = "temperature ~ solar + wind + o3"
model = smf.ols(formula, df).fit()

# 2-1.
print(round(model.params['o3'],3))

# 2-2.
print(round(model.pvalues['wind'],3))

# 2-3.
new_data = pd.DataFrame({'solar': [100], 'wind':[5], 'o3' : [30]})
pred = model.predict(new_data)
print(round(pred[0],3))
```

```python
from statsmodels.formula.api import ols

# 1) R스타일 formula
formula = 'temperature ~ solar + wind + o3'

# 2) 회귀 모델 학습
model= ols(formula, data=df).fit()

# 3) 회귀 모델 요약 정보
summary = model.summary()
print(summary)
```

```python
# 2-1. o3에 대한 회귀계수
print("2-1. o3의 회귀계수:", model.params['o3'])
```

```python
# 2-2. wind에 대한 p-value
print("2-2. wind의 p-value:", model.pvalues['wind'])
```

```python
# 2-3. 예측값
# 1) 새 데이터를 데이터프레임으로 만들기
new_data = pd.DataFrame({
    'solar': [100],
    'wind': [5],
    'o3': [30]
})
print(new_data)
```

```python
# 2) 구축된 모델을 사용해 예측
pred = model.predict(new_data)
print(pred)
```
