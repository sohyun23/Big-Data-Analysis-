### 문제1

제공된 데이터셋(df)의 출동시간과 도착시간 컬럼을 활용하여 출동 후 도착까지 걸린 '시간차이(단위: 분)'를 파생변수로 생성하시오. 이후 소방서별 평균 시간차이를 계산하고, 평균 시간차이가 가장 큰(가장 오래 걸린) 소방서의 평균 시간을 반올림하여 정수 형태로 제출하시오.

```python
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])

df['시간차이(단위: 분)'] = ((df['도착시간']-df['출동시간']).dt.total_seconds())/60

result = df.groupby('소방서')['시간차이(단위: 분)'].mean().max()
print(round(result))
```

```python
import pandas as pd
# df = pd.read_csv("data6-1-1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch6/data6-1-1.csv")

# 1) datetime형태로 변경
df['도착시간'] = pd.to_datetime(df['도착시간'])
df['출동시간'] = pd.to_datetime(df['출동시간'])

# 2) 출동시간과 도착시간의 차이를 분으로 계산
df['시간차이(분)'] = (df['도착시간'] - df['출동시간']).dt.total_seconds() / 60

# 3) 소방서별 평균 시간차이 계산
avg_diff = df.groupby('소방서')['시간차이(분)'].mean()

# 4) 평균 차이가 가장 큰 소방서의 시간을 찾고 출력
idx = avg_diff.idxmax()
result = avg_diff[idx]
print(round(result))
```

```python
# 참고 (시간 반올림)
min = 5.5
print(int(min),"분")
print((min-int(min))*60,"초")
```

### 문제2

```python
df['총학생수'] = df.iloc[:, 2:-1].sum(axis = 1)
df

df['ratio'] = df['총학생수']/ df['교사수']
df['ratio'].idxmax()
print(df.loc[df['ratio'].idxmax()]['교사수'])
```

### 문제3

제공된 데이터셋(df)에서 6가지 종류의 범죄 데이터(2번째부터 7번째 컬럼까지)를 모두 더해 '총 범죄 건수' 파생변수를 생성하시오. 날짜 컬럼에서 '연도(앞 4자리)'를 추출한 뒤 연도별 총 범죄 건수의 합계를 구하시오. 연도별 합계 중 가장 큰 값의 월평균 건수(연도 총합 / 12)를 계산하여 반올림한 정수 형태로 제출하시오.

```python
df['연도'] = df['날짜'].str[:4].astype(int)
result = df.groupby('연도')['총 범죄 건수'].sum().max()/12
print(round(result))
```

```python
import pandas as pd
# df = pd.read_csv("data6-1-3.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch6/data6-1-3.csv")

# 1) 총 범죄 건수 계산
df['총 범죄 건수'] = df.iloc[:, 1:7].sum(axis=1)

# 2) 연도 슬라이싱
df['연도'] = df["날짜"].str[:4]

# 3) 연도별 총 범죄 건수 합 계산
result = df['총 범죄 건수'].groupby(df["연도"]).sum()

# 4) 가장 큰 값의 월 평균 계산
print(round(result.max()/12))
```
