### 문제1

제공된 데이터셋(df)에서 아래 3가지 조건을 모두 만족하는 데이터만 필터링한 후, '2ℓ가격'의 평균값을 구하시오. 계산된 최종 평균값은 반올림하여 정수 형태로 제출하시오.

종량제봉투종류 컬럼의 값이 '규격봉투'

종량제봉투용도 컬럼의 값이 '음식물쓰레기'

2ℓ가격 컬럼의 값이 0이 아닌 데이터

```python
cond1 = df['종량제봉투종류'] == "규격봉투"
cond2 = df['종량제봉투용도'] == "음식물쓰레기"
cond3 = df['2ℓ가격'] != 0

print(round(df[cond1 & cond2 & cond3]['2ℓ가격'].mean()))
```

```python
import pandas as pd
# df = pd.read_csv("data5-1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch5/data5-1.csv")

# 1) 조건1.종량제 봉투 종류 ‘규격봉투’
cond1 = df['종량제봉투종류'] == '규격봉투'

# 2) 조건2.종량제 봉투 용도 ‘음식물쓰레기’
cond2 = df['종량제봉투용도'] == '음식물쓰레기'

# 3) 조건3. 2l가격이 0이면 제외
cond3 = df['2ℓ가격'] != 0

# 4) 조건 적용한 데이터
df = df[cond1 & cond2 & cond3]

# 5) 2l 가격 평균 계산, 반올림, 정수 출력
print(round(df['2ℓ가격'].mean()))
```

### 문제2

제공된 데이터셋(df)의 신장(Height, 단위: cm)과 체중(Weight, 단위: kg) 데이터를 활용하여 BMI 파생변수를 계산하시오. 이후 아래 두 조건에 해당하는 인원수(행 수)의 차이를 절대값으로 구하여 정수 형태로 제출하시오.

(단, $\text{BMI} = \text{체중(kg)} / (\text{신장(m)})^2$ 공식을 사용하시오.)조건 1 (정상체중): $\text{BMI} \ge 18.5$ 이면서 $\text{BMI} < 23$조건 2 (위험체중): $\text{BMI} \ge 23$ 이면서 $\text{BMI} < 25$

```python
df['bmi'] =df['Weight'] /(df['Height']/100)**2
df_normal = df[(df['bmi']>= 18.5) & (df['bmi'] < 23)]
df_abnormal = df[(df['bmi']>= 23) & (df['bmi'] < 25)]

print(abs(len(df_normal) - len(df_abnormal)))

```

```python
import pandas as pd
# df = pd.read_csv("data5-2.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch5/data5-2.csv")

# 1) bmi 파생변수 계산
df['bmi'] = df['Weight'] / (df['Height']/100)**2

# 2) 조건1. 정상체중 bmi
cond1 = (df['bmi'] >= 18.5) & (df['bmi'] < 23)

# 3) 조건2. 위험체중 bmi
cond2 = (df['bmi'] >= 23) & (df['bmi'] < 25)

# 4) 조건1의 인원와 조건2의 인원 차이계산, 절대값 처리
print(abs(len(df[cond1]) - len(df[cond2])))
```

```python
import pandas as pd
# df = pd.read_csv("data5-3.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch5/data5-3.csv")

# 1) '순전입' 파생변수 계산
df['순전입'] = df['전입학생수(계)'] - df['전출학생수(계)']

# 2) '순전입' 컬럼 기준으로 내림차순 정렬
df = df.sort_values('순전입', ascending=False)

# 3) 첫번째 행의 전체 학생 수 값 선택
print(int(df.iloc[0,-2]))
```
