### 문제1-1

제공된 데이터를 바탕으로 Churn을 종속변수로 하고, 나머지 10개 변수 전체를 독립변수로 설정하여 로지스틱 회귀분석을 수행하시오. 분석 결과에서 상수항(Intercept)을 제외하고 유의수준 5%($\alpha = 0.05$) 하에서 통계적으로 유의하지 않은(p-value > 0.05) 독립변수의 총 개수를 구하시오.

```python
import pandas as pd
# df = pd.read_csv("churn.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/churn.csv")
df
```

```python
import statsmodels.formula.api as smf

formula = "Churn ~ AccountWeeks + ContractRenewal + DataPlan + DataUsage + CustServCalls + DayMins + DayCalls + MonthlyCharge + OverageFee + RoamMins "
model = smf.logit(formula, data = df).fit()
print(model.summary())
print((model.pvalues[1:] > 0.05).sum())
```

```python
import pandas as pd
# df = pd.read_csv("churn.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/churn.csv")

# 1) 로지스틱 회귀분석을 위한 포뮬라 생성
formula = "Churn ~ AccountWeeks + ContractRenewal + DataPlan + DataUsage + CustServCalls + DayMins + DayCalls + MonthlyCharge + OverageFee + RoamMins"

# 2) 로지스틱 회귀 모델 생성 및 학습
from statsmodels.formula.api import logit
model = logit(formula, data=df).fit()

# 3) 유의하지 않은 독립변수의 개수 구하기
print(model.summary())
sum(model.pvalues[1:] > 0.05) # Intercept는 독립변수가 아니므로 제외
```

### 문제2-1

제공된 데이터를 바탕으로 PIQ를 종속변수로 하고, Brain, Height, Weight를 독립변수로 설정하여 다중선형회귀분석을 수행하시오. 상수항(Intercept)을 제외한 독립변수 중 p-value(유의확률)가 가장 작은(가장 유의한) 변수의 회귀계수(Coef)를 구하시오.

출력 형식: 소수점 셋째 자리까지 반올림 (예: 2.343)

```python
import pandas as pd
# df = pd.read_csv("piq.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/piq.csv")
df.head()
```

```python
import statsmodels.formula.api as smf
formula = "PIQ ~ Brain + Height + Weight"
model = smf.ols(formula, data =df).fit()
print(model.summary())
idx = model.pvalues[1:].idxmin()

print(round(model.params[idx],3))
```

```python
df.head()
```

```python
# 1) 다중 선형 회귀 분석
from statsmodels.formula.api import ols
model = ols('PIQ ~ Brain + Height + Weight', data=df).fit()

# 2) 가장 작은 p-value를 가진 변수의 회귀계수 찾기
print(model.summary())

# 정답: 2.343
```
