### 문제2

제공된 데이터셋(df)에서 각 행의 관광, 공무, 사업, 기타 컬럼 값을 모두 더하여 '방문객합계' 파생변수를 생성하고, 관광 / 방문객합계 수식을 사용하여 '관광객비율' 파생변수를 추가로 생성하시오. 이후 아래 두 조건에 따라 각각 값을 구한 뒤, 두 값의 합($A + B$)을 계산하여 정수 형태로 제출하시오.

값 A: '관광객비율'을 기준으로 내림차순 정렬했을 때, 상위 2번째 행의 '사업' 방문객 수

값 B: '관광' 방문객 수를 기준으로 내림차순 정렬했을 때, 상위 2번째 행의 '공무' 방문객 수

```python
df['방문객합계'] = df.iloc[:, 1:].sum(axis = 1)
df.head()
```

```python
df['관광객비율'] = df['관광'] / df['방문객합계']
df
```

```python
df2 = df.sort_values(by = '관광객비율', ascending = False)['사업']
A = df2.iloc[1] # 중요

B = df.sort_values(by = '관광', ascending = False)['공무'].iloc[1]
print(A, B, A +B)

```

### 문제3

제공된 데이터셋(df)의 co 컬럼과 nmhc 컬럼에 대해 각각 Min-Max 스케일링(정규화)을 수행하여 0과 1 사이의 값으로 변환하시오. 변환된 두 컬럼의 표준편차를 각각 계산한 후, co 스케일링 값의 표준편차에서 nmhc 스케일링 값의 표준편차를 뺀 차이값을 구하시오. 최종 결과는 소수점 셋째 자리까지 반올림한 실수 형태로 제출하시오.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/chem.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
print(df.head())
df['co'] = scaler.fit_transform(df[['co']])
df['nmhc'] = scaler.fit_transform(df[['nmhc']])
print(df.head())
#print(help(MinMaxScaler))
```

```python
std_co = df['co'].std()
std_nmhc = df['nmhc'].std()
result = std_co- std_nmhc
print(round(result, 3))
```

```python
# 방법1
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# df = pd.read_csv("chem.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/chem.csv")

# 1) Min-Max 스케일링 (데이터프레임으로 만들기)
scaler = MinMaxScaler()
df['co_scaled'] = scaler.fit_transform(df[['co']])
df['nmhc_scaled'] = scaler.fit_transform(df[['nmhc']])

# 2) 표준편차 계산 (판다스 표준편차)
co_std = df['co_scaled'].std()
nmhc_std = df['nmhc_scaled'].std()
print(co_std, nmhc_std)

# 3) 표준편차 차이 계산 및 반올림
std_diff = round(co_std - nmhc_std, 3)
print(std_diff)
```

```python
# 방법2
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# df = pd.read_csv("chem.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/chem.csv")

# 1) Min-Max 스케일링 (transform의 결과는 넘파이)
scaler = MinMaxScaler()
co_scaled = scaler.fit_transform(df[['co']])
nmhc_scaled = scaler.fit_transform(df[['nmhc']])

# 2) 표준편차 계산 (넘파이 표준편차)
co_std = co_scaled.std()
nmhc_std = nmhc_scaled.std()
print(co_std, nmhc_std)

# 3) 표준편차 차이 계산 및 반올림
std_diff = round(co_std - nmhc_std, 3)
print(std_diff)
```
